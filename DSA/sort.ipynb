```python
####linear search(start from starting point and go till end start from 0) 
##time com: 0(N)
##no sorting required
```

```python
def linear(arr,item):
    for i in range(len(arr)):
        if arr[i]==item:
                return i
        return -1
```

```python
arr=[1,2,4,77,88,4]
linear(arr,88)
```

```python
def binarys(arr,low,high,item):
    arr.sort()  
    if low<=high:
        mid=(low+high)//2+
        if arr[mid]==item:
            return mid
        elif arr[mid]>item:
            return binarys(arr,low,mid-1,item)
        elif arr[mid]<item:
            return binarys(arr,mid+1,high,item)
    else:
        return -1
```

```python
def sort(arr):
    sorted=True
    for i in range(len(arr)-1):
        if arr[i]> arr[i+1]:
            sorted=False
    return sorted
```

```python
import threading
import time

def sleep_sort(numbers):
    """
    Sleep Sort: Sort numbers by leveraging delays proportional to their values.

    Args:
    numbers (list): List of non-negative numbers to sort.

    Returns:
    None: Prints the sorted numbers as they are sorted.
    """
    def sleep_and_print(n):
        time.sleep(n)  # Delay proportional to the number
        print(n, end=" ")  # Print the number after the delay

    threads = []
    for num in numbers:
        thread = threading.Thread(target=sleep_and_print, args=(num,))
        threads.append(thread)
        thread.start()

    for thread in threads:
        thread.join()  # Ensure all threads complete before exiting


```

```python
# Example usage
numbers = [3, 1, 4, 1, 5, 9]
print("Sorted Output:", end=" ")
sleep_sort(numbers)
```

```python
def merge(arr1,arr2):
    i=j=0
    merged=[]
    while i < len(arr1) and j <len(arr2):
        if arr1[i]<arr2[j]:
            merged.append(arr1[i])
            i+=1
        else:
            merged.append(arr2[j])
            j+=1
    while i<len(arr1):
        merged.append(arr1[i])
        i+=1
    while j<len(arr2):
        merged.append(arr2[j])
        j+=1
    return merged
```

```python
def merged(arr):
    if len(arr)==1:
        return arr
    mid =len(arr)//2
    left =arr[:mid]
    right=arr[mid:]
    left = merged(left)
    right=merged(right)
    return merge(left,right)
```

```python
def insertion(arr1):
    n=len(arr1)
    for i in range (1,n):
        key=arr1[i]
        j=i-1
        while j>=0 and key< arr1[j]:
            arr1[j+1]=arr1[j]
            j-=1
        arr1[j+1]=key
    return arr1
```

```python
arr1=[3,2,5,6,8]
```

```python
insertion(arr1)
```

```python
def quick_sort(arr, low, high):
    if low < high:
        # Get the pivot index after partition
        pivot = partition(arr, low, high)
        # Recursively sort the left and right partitions
        quick_sort(arr, low, pivot - 1)
        quick_sort(arr, pivot + 1, high)
        return arr

def partition(arr, low, high):
    pivot = arr[low]  # Choose the first element as the pivot
    i = low + 1
    j = high

    while True:
        # Move i to the right until an element larger than pivot is found
        while i <= j and arr[i] <= pivot:
            i += 1
        # Move j to the left until an element smaller than pivot is found
        while i <= j and arr[j] >= pivot:
            j -= 1

        if i <= j:
            # Swap elements at i and j
            arr[i], arr[j] = arr[j], arr[i]
        else:
            break

    # Swap the pivot element with the element at j
    arr[low], arr[j] = arr[j], arr[low]
    return j  # Return the pivot index


```

```python
def heapify(arr, n, i):
    largest = i  
    left = 2 * i + 1  # Left child
    right = 2 * i + 2  # Right child

    # Check if left child exists and is greater than root
    if left < n and arr[left] > arr[largest]:
        largest = left

    # Check if right child exists and is greater than root
    if right < n and arr[right] > arr[largest]:
        largest = right

    # If largest is not root
    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]  # Swap
        # Recursively heapify the affected subtree
        heapify(arr, n, largest)


def heap_sort(arr):
    n = len(arr)

    # Build a max heap
    for i in range(n // 2 - 1, -1, -1):
        heapify(arr, n, i)

    # Extract elements one by one
    for i in range(n - 1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]  # Swap
        heapify(arr, i, 0)

```

```python

# Example usage
arr = [12, 11, 13, 5, 6, 7]
print("Original array:", arr)
heap_sort(arr)
print("Sorted array:", arr)
```
