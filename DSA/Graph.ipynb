```python
class Graph:
    def __init__(self, edges):
        self.edges = edges
        self.graph_dict = {}
        for start, end in self.edges:
            if start in self.graph_dict:
                self.graph_dict[start].append(end)
            else:
                self.graph_dict[start] = [end]
        print("Graph:", self.graph_dict)

    def get_path(self, start, end, path=[]):
        path = path + [start]  # Add the current node to the path
        if start == end:
            return [path]  # Return the path as a list of paths
        if start not in self.graph_dict:
            return []  # No path if the start node is not in the graph

        total_paths = []
        for node in self.graph_dict[start]:
            if node not in path:  # Avoid revisiting nodes
                new_paths = self.get_path(node, end, path)  # Recursive call
                total_paths.extend(new_paths)  # Add new paths to total_paths

        return total_paths

    def short(self, start, end, path=[]):
        path = path + [start]  # Add the current node to the path
        if start == end:
            return path  # Return the current path as it's complete

        if start not in self.graph_dict:
            return None  # No path if the start node is not in the graph

        shortest = None
        for node in self.graph_dict[start]:
            if node not in path:  # Avoid revisiting nodes
                sp = self.short(node, end, path)  # Recursive call
                if sp:
                    if shortest is None or len(sp) < len(shortest):  # Compare lengths
                        shortest = sp

        return shortest


if __name__ == '__main__':
    routes = [
        ("mumbai", "paris"),
        ("mumbai", "Dubai"),
        ("paris", "Dubai"),
        ("paris", "new york"),
        ("Dubai", "new york"),
        ("new york", "use")
    ]

    route_graph = Graph(routes)
    start = "mumbai"
    end = "new york"

    print(f"Paths between {start} and {end}: {route_graph.get_path(start, end)}")
    print()
    print(f"Shortest path between {start} and {end}: {route_graph.short(start, end)}")
```
