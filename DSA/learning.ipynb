```python
# property decorator
class employee:
    a=1
    @classmethod
    def show(cls):
        print(f"The class value of a is {cls.a}")
    @property
    def name (self):
        return f"{self.fname} {self.iname}"
    @name.setter
    def name(self, value): 
        self.fname=value.split(" ")[0]
        self.iname=value.split(" "[0])
e=employee()
e.a=45
e.name="harry sharma"
print(e.name)
e.show()
```

```python
class Complex:
    def __init__(self, r, i):
        self.r = r
        self.i = i

    def __add__(self, c2):
        return Complex(self.r + c2.r, self.i + c2.i)

    def __mul__(self, c2):
        real_part = self.r * c2.r - self.i * c2.i
        imaginary_part = self.r * c2.i + self.i * c2.r
        return Complex(real_part, imaginary_part)

    def __str__(self):
        return f"{self.r} + {self.i}i"

# Example usage
c1 = Complex(1, 2)
c2 = Complex(3, 4)

# Print the result of addition
print("Addition:", c1 + c2)  # Output: 4 + 6i

# Print the result of multiplication
print("Multiplication:", c1 * c2)  # Output: -5 + 10i
```
