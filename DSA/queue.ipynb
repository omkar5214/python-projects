```python
class node:
    def __init__(self,value):
        self.data=value
        self.next=None
```

```python
class queue:
    def __init__(self):
        self.front = None
        self.rear = None
        
    def enqueue(self, value):
        new_node = Node(value)
        if self.rear is None:
            self.front = new_node
            self.rear = new_node
        else:
            self.rear.next = new_node
            self.rear = new_node
    
    def dequeue(self):
        if self.front is None:
            return "empty"
        else:
            dequeued_value = self.front.data
            self.front = self.front.next
            if self.front is None:
                self.rear = None
            return dequeued_value
        
    def isempty(self):
        return self.front==None
    def size(self):
        temp=self.front
        counter=0
        while temp!=None:
            counter+=1
            temp=temp.next
        return counter
    def fpeak(self):
        if self.front==None:
            return "empty"
        else:
            return self.front.data
        
    def tpeak(self):
        if self.front==None:
            return "empty"
        else:
            return self.rear.data
    
    def traverse(self):
        temp = self.front
        while temp is not None:
            print(temp.data,end=" ")
            temp = temp.next
```

```python
q=queue()
```

```python
q.enqueue(4)
q.enqueue(3)
q.enqueue(6)
```
